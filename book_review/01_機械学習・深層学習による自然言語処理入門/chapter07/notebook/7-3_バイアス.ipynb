```python
import matplotlib.pyplot as plt
import numpy as np
```

### バイアスなし

```python
x = np.arange(-6, 6, 0.1)
weight = [0.5, 1.0, 2.0]

for i, w in enumerate(weight):
    f = 1 / (1 + np.exp(-x * w))
    label = 'w{} = {}'.format(i, w)
    plt.plot(x, f, label=label)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc=2)
plt.show()
```

### バイアスあり

```python
x = np.arange(-6, 6, 0.1)
w = 5
biases = [-8.0, 0.0, 8.0]

for i, b in enumerate(biases):
    f = 1 / (1 + np.exp(-x * w + b))
    label = 'b{} = {}'.format(i, b)
    plt.plot(x, f, label=label)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc=2)
plt.show()
```
